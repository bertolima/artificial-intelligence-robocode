```python
import numpy as np
%pip install scikit-fuzzy
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
fp_distance = np.arange(0, 2000, 1)  
fp_enemyEnergy = np.arange(0, 200, 1)  
fp_energy = np.arange(0, 200, 1)  
fp_firepower = np.arange(0.0, 30.0, 1)

distance_near = fuzz.trapmf(fp_distance,[0, 0, 150, 300])  
distance_reg = fuzz.trapmf(fp_distance,[175, 250, 450, 600])  
distance_far = fuzz.trapmf(fp_distance,[475, 550, 2000, 2000]) 

enemyEnergy_low = fuzz.trapmf(fp_enemyEnergy,[0, 0, 20, 40])  
enemyEnergy_reg = fuzz.trapmf(fp_enemyEnergy,[25, 40, 60, 70])  
enemyEnergy_high = fuzz.trapmf(fp_enemyEnergy,[60, 70, 200, 200])  
	  
energy_low = fuzz.trapmf(fp_energy,[0, 0, 20, 40])  
energy_reg = fuzz.trapmf(fp_energy,[25, 40, 60, 70])  
energy_high = fuzz.trapmf(fp_energy,[60, 70, 200, 200])  

fp_low = fuzz.trapmf(fp_firepower,[0, 0, 5, 15])  
fp_reg = fuzz.trapmf(fp_firepower,[10, 15, 20, 25])  
fp_high = fuzz.trapmf(fp_firepower,[20, 25, 30, 30])  
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(8, 5))


ax0.plot(fp_distance, distance_near, 'b', linewidth=1.5, label='near')  
ax0.plot(fp_distance, distance_reg, 'g', linewidth=1.5, label='regular')  
ax0.plot(fp_distance, distance_far, 'r', linewidth=1.5, label='far')  
ax0.set_title('Distância do Robô Inimigo')  
ax0.legend() 

ax1.plot(fp_enemyEnergy, enemyEnergy_low, 'b', linewidth=1.5, label='low')  
ax1.plot(fp_enemyEnergy, enemyEnergy_reg, 'g', linewidth=1.5, label='regular')  
ax1.plot(fp_enemyEnergy, enemyEnergy_high, 'r', linewidth=1.5, label='high')  
ax1.set_title('Energia do Robô Inimigo')  
ax1.legend()  
	  

for ax in (ax0, ax1):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()

```

```python
fig2, (ax2, ax3) = plt.subplots(nrows=2, figsize=(8, 5))

ax2.plot(fp_energy, energy_low, 'b', linewidth=1.5, label='low')  
ax2.plot(fp_energy, energy_reg, 'g', linewidth=1.5, label='regular')  
ax2.plot(fp_energy, energy_high, 'r', linewidth=1.5, label='high')  
ax2.set_title('Energia do Robô')  
ax2.legend()

ax3.plot(fp_firepower, fp_low, 'b', linewidth=1.5, label='weak')  
ax3.plot(fp_firepower, fp_reg, 'g', linewidth=1.5, label='regular')  
ax3.plot(fp_firepower, fp_high, 'r', linewidth=1.5, label='strong')  
ax3.set_title('Força do Tiro (10x)')  
ax3.legend()

for ax in (ax2, ax3):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()
```

```python

gd_enemyEnergy = np.arange(0, 200, 1)  
gd_energy = np.arange(0, 200, 1)  
gd_distance = np.arange(80, 600, 1)  

gd_enemyEnergy_low = fuzz.trapmf(gd_enemyEnergy,[0, 0, 20, 40])  
gd_enemyEnergy_reg = fuzz.trapmf(gd_enemyEnergy,[25, 40, 60, 70])  
gd_enemyEnergy_high = fuzz.trapmf(gd_enemyEnergy,[60, 70, 200, 200])  
	  
gd_energy_low = fuzz.trapmf(gd_energy,[0, 0, 20, 40])  
gd_energy_reg = fuzz.trapmf(gd_energy,[25, 40, 60, 70])  
gd_energy_high = fuzz.trapmf(gd_energy,[60, 70, 200, 200])

gd_distance_near = fuzz.trapmf(gd_distance,[80, 150, 180, 230])  
gd_distance_reg = fuzz.trapmf(gd_distance,[190, 240, 270, 350])  
gd_distance_far = fuzz.trapmf(gd_distance,[300, 400, 500, 600]) 
```

```python
fig2, (gd1, gd2, gd3) = plt.subplots(nrows=3, figsize=(8, 5))


gd1.plot(gd_enemyEnergy, gd_enemyEnergy_low, 'b', linewidth=1.5, label='low')  
gd1.plot(gd_enemyEnergy, gd_enemyEnergy_reg, 'g', linewidth=1.5, label='regular')  
gd1.plot(gd_enemyEnergy, gd_enemyEnergy_high, 'r', linewidth=1.5, label='high')  
gd1.set_title('Distância do Robô Inimigo')  
gd1.legend() 

gd2.plot(gd_energy, gd_energy_low, 'b', linewidth=1.5, label='low')  
gd2.plot(gd_energy, gd_energy_reg, 'g', linewidth=1.5, label='regular')  
gd2.plot(gd_energy, gd_energy_high, 'r', linewidth=1.5, label='high')  
gd2.set_title('Energia do Robô')  
gd2.legend()

gd3.plot(gd_distance, gd_distance_near, 'b', linewidth=1.5, label='near')  
gd3.plot(gd_distance, gd_distance_reg, 'g', linewidth=1.5, label='regular')  
gd3.plot(gd_distance, gd_distance_far, 'r', linewidth=1.5, label='far')  
gd3.set_title('Distancia Alvo')  
gd3.legend() 
	  

for ax in (gd1, gd2, gd3):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()
```

```python

ga_speed = np.arange(0.0, 80, 1, dtype='float64')
ga_distance = np.arange(0, 2000, 1) 
ga_angulation = np.arange(30, 150, 1)  

ga_speed_low = fuzz.trimf(ga_speed,[0.0, 20, 40])  
ga_speed_reg = fuzz.trimf(ga_speed,[30, 45, 55])  
ga_speed_high = fuzz.trapmf(ga_speed,[45, 60, 80, 80])

ga_distance_near = fuzz.trapmf(ga_distance,[0, 0, 150, 300])  
ga_distance_reg = fuzz.trapmf(ga_distance,[200, 300, 400, 500])  
ga_distance_far = fuzz.trapmf(ga_distance,[450, 550, 2000, 2000]) 
	  
ga_angulation_low = fuzz.trapmf(ga_angulation,[45, 60, 75, 90])  
ga_angulation_lot = fuzz.trapmf(ga_angulation,[70, 90, 105, 120])


```

```python
fig3, (ga1, ga2, ga3) = plt.subplots(nrows=3, figsize=(8, 5))


ga1.plot(ga_speed, ga_speed_low, 'b', linewidth=1.5, label='slow')  
ga1.plot(ga_speed, ga_speed_reg, 'g', linewidth=1.5, label='regular')  
ga1.plot(ga_speed, ga_speed_high, 'r', linewidth=1.5, label='fast')  
ga1.set_title('Velocidade do Robô Inimigo (10x)')  
ga1.legend() 

ga2.plot(ga_distance, ga_distance_near, 'b', linewidth=1.5, label='near')  
ga2.plot(ga_distance, ga_distance_reg, 'g', linewidth=1.5, label='regular')  
ga2.plot(ga_distance, ga_distance_far, 'r', linewidth=1.5, label='far')  
ga2.set_title('Distancia do Robô Inimigo')  
ga2.legend()

ga3.plot(ga_angulation, ga_angulation_low, 'b', linewidth=1.5, label='low')  
ga3.plot(ga_angulation, ga_angulation_lot, 'g', linewidth=1.5, label='lot')  
ga3.set_title('Angulação de Saída')  
ga3.legend() 
	  

for ax in (ga1, ga2, ga3):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()
```
